```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
```

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data,columns = cancer.feature_names)
df.head()
```

```python
X = df  # training feature
Y =  pd.Series(cancer.target)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.20,shuffle= True,random_state=2)
SVC_model = svm.SVC()
SVC_model.fit(X_train, Y_train)
```

```python
KNN_model = KNeighborsClassifier(n_neighbors=5)
KNN_model.fit(X_train, Y_train)
```

```python
SVC_prediction = SVC_model.predict(X_test)
```

```python
print("Accuracy :",np.round(accuracy_score(Y_test,SVC_prediction),2))
precision,recall,fscore,_ = precision_recall_fscore_support(Y_test,SVC_prediction,average = 'binary')
```

```python
print("Precision :",np.round(precision,2))
print("Recall :",np.round(recall,2))
print("fscore :",np.round(fscore,2))
```

```python
# Make the confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(Y_test, SVC_prediction)
print("\nConfusion Matrix")
sns.heatmap(cf_matrix, annot=True, cmap='Blues')
plt.xlabel('Predicted', fontsize=12)
plt.ylabel('True', fontsize=12)
```

```python
KNN_predict = KNN_model.predict(X_test)
```

```python
print("Accuracy :",np.round(accuracy_score(Y_test,KNN_predict),2))
precision,recall,fscore,_ = precision_recall_fscore_support(Y_test,KNN_predict,average = 'binary')
```

```python
print("Precision :",np.round(precision,2))
print("Recall :",np.round(recall,2))
print("fscore :",np.round(fscore,2))
```

```python
# Make the confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(Y_test, KNN_predict)
print("\nConfusion Matrix")
sns.heatmap(cf_matrix, annot=True, cmap='Blues')
plt.xlabel('Predicted', fontsize=12)
plt.ylabel('True', fontsize=12)
```
